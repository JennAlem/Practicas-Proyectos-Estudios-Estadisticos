#### Importar bibliotecas relevantes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Cargar Base de Datos

```python
restaurante = pd.read_excel("Restaurante.xlsx")
```

```python
restaurante.head()
```

## Subconjunto de columnas relevantes

```python
restaurante_data = restaurante[['Atendió', 'Categoria', 'Precio', 'Costo']]
```

## Inspeccionar las primeras 5 filas y los tipos de datos del conjunto de datos

```python
restaurante_data.head()
```

```python
restaurante_data.shape
```

```python
restaurante_data.dtypes
```

## Crear una variable de margen de beneficio en función de las variables de precio y costo.

```python
restaurante_data['Margen de beneficio'] = (restaurante_data['Precio'] - restaurante_data['Costo']) / restaurante_data['Precio'] * 100
```

## Ordenar casas según precios

```python
restaurante_sorted = restaurante_data.sort_values('Precio', ascending = False)
```

```python
restaurante_sorted.head()
```

## Trazar un gráfico de barras usando Seaborn

```python
plt.figure(figsize= (12,6))
data = restaurante_sorted[55:70]
sns.barplot(data= data, x= 'Atendió',y = 'Precio')
```

## Agregue otros detalles para que el cuadro sea más informativo.

```python
plt.figure(figsize= (12,6))
data = restaurante_sorted[55:75]

ax = sns.barplot(data= data, x= 'Atendió',y = 'Precio')
ax.set_xlabel('Personal code',fontsize = 15)
ax.set_ylabel('Ventas por personal', fontsize = 15)
ax.set_title('Top 10 Personal', fontsize= 20)
```

## Utilice métodos de Matplotlib para obtener detalles del gráfico

```python
plt.figure(figsize= (12,6))
sns.barplot(data= data, x= 'Atendió',y = 'Precio')
plt.title('Top 10 Personal', fontsize=15)
plt.xlabel('Personal code', fontsize = 12)
plt.xticks(fontsize=10)
plt.ylabel('Ventas por personal', fontsize=12)
plt.yticks(fontsize=10)
plt.show()
```

## Crear Subplots

```python
fig, ax = plt.subplots(1, 2,figsize=(40,18))

data = restaurante_sorted[0:10]

sns.set(font_scale = 3)
ax1 = sns.barplot(data= data, x= 'Atendió',y = 'Precio', ax = ax[0])
ax1.set_xlabel('Personal code')
ax1.set_ylabel('Ventas por personal')
ax1.set_title('Top 10 Personal')


ax2 = sns.barplot(data= data, x= 'Atendió',y = 'Margen de beneficio', ax=ax[1])
ax2.set_xlabel('Ventas por personal code')
ax2.set_ylabel('Margen de beneficio generado por las ventas de personal')
ax2.set_title('Top 10 Personal')
```
