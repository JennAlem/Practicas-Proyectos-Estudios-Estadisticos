# Importar librerias importantes

```python
import pandas as pd
from bokeh.plotting import figure, show
import bokeh.plotting as bk_plot
from bokeh.io import output_notebook
output_notebook()
```

# Cargar la base de datos

```python
restaurante = pd.read_excel("Restaurante.xlsx")
```

```python
restaurante.head()
```

#Columnas Relevantes

```python
restaurante_data = restaurante[['Atendió','Producto','Precio', 'Costo', 'Tipo de Cliente']]
```

# Inspeccionar las primeras filas de la base

```python
restaurante_data.head()
```

```python
restaurante_data.shape
```

```python
restaurante_data.dtypes
```

# Crear una variable de margen de beneficio = (Precio de venta - Costo del producto) / Precio de venta * 100

```python
restaurante_data['Margen de beneficio'] = (restaurante_data['Precio'] - restaurante_data['Costo']) / restaurante_data['Precio'] * 100
```

# Ordenar Productos según Precios

```python
restaurante_data = restaurante_data.sort_values('Precio', ascending = False)
```

```python
restaurante_data.head()
```

```python
restaurante_data.drop([30132,], axis=0, inplace=True)
```

```python
restaurante_data.head()
```

```python
restaurante_data.iloc[2:6, :]
```

# Trazar un gráfico de barras usando Bokeh

```python
data = restaurante_data[2:6]

fig = figure(x_range = data['Atendió'])
fig.vbar(x= data['Atendió'], top = data['Precio'], width = 0.9)
show(fig)
```

# **Agregue otros detalles para que el cuadro sea más informativo.**

```python
fig = figure(x_range = data['Atendió'],
             title = 'Top 10 de empleados', x_axis_label = 'Empleado',
             y_axis_label = 'Margen de beneficio')

fig.vbar(x= data['Atendió'], top = data['Margen de beneficio'], width = 0.9)


fig.xaxis.axis_label_text_font_size = "15pt"
fig.xaxis.major_label_text_font_size = "10pt"
fig.yaxis.axis_label_text_font_size = "15pt"
fig.yaxis.major_label_text_font_size = "10pt"
fig.title.text_font_size = '15pt'


show(fig)
```

# **Crear Subplots**

```python
p1 = figure(x_range = data['Atendió'],
             title = 'Top 10 ', x_axis_label = 'Personal',
             y_axis_label = 'Margen de beneficio')

p1.vbar(x= data['Atendió'], top = data['Precio'], width = 0.9)

p2 = figure(x_range = data['Atendió'],
             title = 'Top 10', x_axis_label = 'Personal',
             y_axis_label = 'Precio')

p2.vbar(x= data['Atendió'], top = data['Margen de beneficio'], width = 0.9)


gp = bk_plot.gridplot(children=[[p1, p2]])
bk_plot.show(gp)
```
